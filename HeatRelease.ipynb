```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df2 = pd.read_excel('HeatRelease.xlsx', sheet_name='Tabelle1')
```

```python
df2.head()
```

```python
df2 = df2.rename(columns={'time' : 'time [h]'})
df2 = df2.rename(columns={'Unnamed: 1' : 'd'})
```

```python
df2.drop(df2.index[0], inplace=True)
df2.head()
```

```python
df2.info()
```

```python
df2.describe()
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df2['d'], df2['heattotal_EVA_15'], color='orange', label='EVA_15')
ax.plot(df2['d'], df2['heattotal_SA_15'], color='red', label='SA_15')
ax.plot(df2['d'], df2['heattotal_SBR_15'], color='blue', label='SBR_15')
ax.plot(df2['d'], df2['heattotal_REF'], color='green', label='REF')
ax.set_xlabel('Time [d]')
ax.set_ylabel('Heat Total')
ax.legend()
plt.savefig('Second Sheet\HeatRelease.png')
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df2['d'], df2['PCS_EVA15 (20C)'], color='orange', label='PCS_EVA15 (20C)')
ax.plot(df2['d'], df2['PCS_SA15 (20C)'], color='red', label='PCS_SA15 (20C)')
ax.plot(df2['d'], df2['PCS_SBR15 (20C)'], color='blue', label='PCS_SBR15 (20C)')
ax.plot(df2['d'], df2['PCS_REF (20C)'], color='green', label='PCS_REF (20C)')
ax.set_xlabel('Time [d]')
ax.set_ylabel('PCS')
ax.legend()
plt.savefig('Second Sheet\pcs_20C.png')
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df2['d'], df2['PCS_EVA15 (60C)'], color='orange', label='PCS_EVA15 (60C)')
ax.plot(df2['d'], df2['PCS_SA15 (60C)'], color='red', label='PCS_SA15 (60C)')
ax.plot(df2['d'], df2['PCS_SBR15 (60C)'], color='blue', label='PCS_SBR15 (60C)')
ax.plot(df2['d'], df2['PCS_REF (60C)'], color='green', label='PCS_REF (60C)')
ax.set_xlabel('Time [d]')
ax.set_ylabel('PCS')
ax.legend()
plt.savefig('Second Sheet\pcs_60C.png')
```
